````python
import time
import numpy as np
import matplotlib.pyplot as plt
import lenstest

def draw_circle(R,X0=0,Y0=0,color='black'):
    """Draw a circle."""
    theta = np.linspace(0, 2*np.pi, 100)
    plt.plot(X0+R*np.sin(theta), Y0+R*np.cos(theta), color=color)
````

````python
def knife_polygon(side, phi, dx):
    """
    Create a polygon for a rotated knife edge.
    
    The polygon is a 1:2 rectangle that has been rotated by an
    angle phi from the vertical and shifted by a distance dx.
    The 

    Args:
        side: length of the short side of the rectangle [mm]
        phi: rotation from vertical (positive ==> CCW) [radians]
        dx: horizontal offset [mm]
    Returns:
        x, y: coordinates of polygon
    """
    r = 1.5 * side
    rad = phi + np.pi

    x = np.full(5, (dx + 2 * r) * np.cos(rad))
    y = np.full(5, (dx + 2 * r) * np.sin(rad))
    
    rad = rad + np.pi / 2
    x[1] = x[0] + 1 * r * np.cos(rad)
    y[1] = y[0] + 1 * r * np.sin(rad)

    rad = rad + np.pi / 2
    x[2] = x[1] + 2 * r * np.cos(rad)
    y[2] = y[1] + 2 * r * np.sin(rad)

    rad = rad + np.pi / 2
    x[3] = x[2] + 1 * r * np.cos(rad)
    y[3] = y[2] + 1 * r * np.sin(rad)

    return x,y

def knife_polygon2(s, phi, ds):
    """
    Create a polygon for a rotated knife edge.
    
    The polygon is a vertical 1:2 rectangle that is shifted and
    then rotated counter clockwise.  
    
    When the shift is zero, then the edge of the knife edge is at the origin.
    Specifing a shift translates the knife edge across the origin.

    Args:
        s: short side of the rectangle [mm]
        phi: CCW rotation              [radians]
        ds: shift of knife edge        [mm]
    Returns:
        x, y: coordinates of polygon
    """
    # vertical s x 2s rectangle with center of edge shifted by dx
    xp = np.array([0,s,s,0,0]) + ds
    yp = np.array([s,s,-s,-s,s])
    
    # rotate rectangle CCW
    x = xp * np.cos(phi) - yp * np.sin(phi)
    y = xp * np.sin(phi) + yp * np.cos(phi)
    
    return x,y
````

````python
dx=50
x, y = knife_polygon2(100, np.radians(45), dx)
plt.gca().set_aspect('equal')
plt.fill(x, y, color='black', alpha=0.8)
plt.xlabel("Knife Edge Plane (mm)")
plt.plot(0,0,'bo')

plt.show()
````

## Original Source

From the April 1991 issue of *Sky and Telescope*, page 419, has code written in BASIC that creates 
a ronchigram for a parabolic mirror.

https://skyandtelescope.org/astronomy-resources/basic-programs-from-sky-telescope/

Here is a listing of the original basic program.  It is only 51 lines long and was pretty easy to port
to python.

```
    10 REM       RONCHI.BAS
    20 REM
    30 INPUT "Mirror diameter    ";D
    40 INPUT "Radius of curvature";R
    50 INPUT "Grating frequency  ";F
    60 PRINT "Grating distance Delta"
    70 PRINT "  from the mirror's 
    80 PRINT "  center of curvature"
    90 INPUT "  (+ is outside)   ";DL
    100 W=1/(2*F): REM  Line width
    120 CLS
    130 SCREEN 2
    140 X0=300: Y0=100: C=300
    150 K=.42
    160 CIRCLE (X0,Y0),C/2
    170 FOR I=1 TO 10000
    180 X=D*(RND(1)-.5)
    190 Y=D*(RND(1)-.5)
    200 REM   X and Y are the ray's
    210 REM   coordinates on the face
    220 REM   of the mirror
    230 S2=X*X+Y*Y
    240 IF SQR(S2)>D/2 THEN 390
    250 Z=R+S2/R
    260 L=R+DL-Z
    270 U=L*X/Z
    280 REM   Now, test to see if the
    290 REM   ray is blocked (FL=0)
    300 REM   or transmitted (FL=1)
    310 REM   by the grating
    320   FL=0: REM  Reset flag
    330   T=INT(ABS(U/W)+.5)
    340   IF T/2=INT(T/2) THEN FL=1
    350   IF FL=0 THEN 390
    360 XP=X0+X*C/D
    370 YP=Y0+Y*K*C/D
    380 PSET(XP,YP): REM  Plot point
    390 NEXT I
    400 LOCATE 1,1
    410   PRINT "Diameter =   ";D
    420 LOCATE 2,1
    430   PRINT "R of C =     ";R
    440 LOCATE 3,1
    450   PRINT "Ronchi freq =";F
    460 LOCATE 4,1
    470   PRINT "Delta =      ";DL
    480 END
    490 REM ************************************
    500 REM   APPEARED IN ASTRONOMICAL COMPUTING
    510 REM   SKY & TELESCOPE - APRIL 1991 ISSUE
    520 REM ************************************
```

## First translation of the BASIC program

This version is slow.

````python
start = time.time()

Fnumber = 4
D = 10     # inches mirror diameter
F = 100    # lines per inch (grating frequency)
DL = -0.62 # Grating to mirrors center
R = 160    # inches

W=1/(2*F)  # lines per inch

X0=0
Y0=0
C=300
K=0.42
Rsqr = (D/2)**2
K=1

print("           Diameter %6.2f in" % D)
print("Radius of Curvature %6.2f in" % R)
print("              Delta %6.2f in" % DL)
print("       Focal Length %6.2f in" % (R/2))
print("                 F# %6.2f" % (R/2/D))
print("        Ronchi freq %6.2f lines per in" % F)
print()

draw_circle(R,X0,Y0)

for i in range(10000):
    # X,Y are coordinates on the face of the mirror
    X=D*(np.random.random()-0.5)
    Y=D*(np.random.random()-0.5)
    S2=X*X+Y*Y
    if S2 <= Rsqr:
        Z=R+S2/R
        L=R+DL-Z
        U=L*X/Z

# Now, test to see if the ray is blocked (FL=0) or transmitted (FL=1) by the grating
        T=int(abs(U/W)+.5)
        if T % 2 > 0:
            XP=X0+X*C/D
            YP=Y0+Y*K*C/D
            plt.plot([XP],[YP],'o', markersize=1, color='black')

plt.gca().set_aspect('equal')
plt.show()
end = time.time()
print("elapsed time = %.0f ms" % (1000*(end - start)))
````

## Revised version without loops

This version is some 30X faster and does 10X more points.

````python
start = time.time()

Fnumber = 4
D = 10     # inches mirror diameter
F = 100    # lines per inch (grating frequency)
DL = -0.62 # Grating to mirrors center
R = 160    # inches

W=1/(2*F)  # lines per inch

npoints = 100000
X0=300
Y0=100
C=300
K=0.42

K=1 

print("           Diameter %6.2f in" % D)
print("Radius of Curvature %6.2f in" % R)
print("              Delta %6.2f in" % DL)
print("       Focal Length %6.2f in" % (R/2))
print("                 F# %6.2f" % (R/2/D))
print("        Ronchi freq %6.2f lines per in" % F)
print()


plt.subplots(1,1,figsize=(6,4))
plt.subplot(1,1,1)
draw_circle(R,X0,Y0)

U1 = np.random.uniform(size = npoints)
U2 = np.random.uniform(size = npoints)
X = D/2 * np.sqrt(U2) * np.cos(2 * np.pi * U1)
Y = D/2 * np.sqrt(U2) * np.sin(2 * np.pi * U1)

Z=R+(X**2+Y**2)/R

U=(R+DL)*X/Z - X

T=(np.abs(U/W)+0.5).astype(int)
Tmask = T%2==1
XP=X0+X*C/D
YP=Y0+Y*K*C/D

xp_mask = np.ma.masked_where(Tmask, XP)
yp_mask = np.ma.masked_where(Tmask, YP)

plt.plot(xp_mask,yp_mask,'o', markersize=0.1, color='black')

plt.gca().set_aspect('equal')
plt.show()
end = time.time()
print("elapsed time = %.0f ms" % (1000*(end - start)))
````

## Ronchigrams passing through focus

Here are six ronchigrams at different locations inside and outside the focus.

````python
start = time.time()

mirror_diameter = 20*25.4  # mirror diameter
radius_of_curvature = 160 * 25.4

lp_per_mm = 100/25.4  # grating frequency lp/mm
offset_from_focus = -0.47*25.4  # Grating distance Delta from the mirrors center of curvature
# positive values are outside radius of curvature

line_width = 0.5/lp_per_mm

X0 = 1500
Y0 = 0
C = 300
K = 0.42
mirror_radius = mirror_diameter/2
focal_length = radius_of_curvature/2
f_number = focal_length/mirror_diameter

K=1

print("    Mirror Diameter = %.0f mm" % mirror_diameter)
print("Radius of Curvature = %.0f mm" % radius_of_curvature)
print("   Ronchi Frequency = %.0f lp/mm" % lp_per_mm)
print("       Focal Length = %.0f mm" % focal_length)
print("                 F# = %.1f" % f_number)


plt.subplots(2,3,figsize=(10,6))


for i, offset in enumerate([-10,-5,0,5,10,15]):
    plt.subplot(2,3,i+1)
    offset_from_focus = offset
    
    # draw a circle
    theta = np.linspace(0, 2*np.pi, 100)
    plt.gca().set_aspect('equal')
    plt.plot(X0+C/2*np.sin(theta), Y0+C/2*np.cos(theta))

    npoints = 100000
    U1 = np.random.uniform(size=npoints)
    U2 = np.random.uniform(size=npoints)
    X = mirror_radius * np.sqrt(U2) * np.cos(2 * np.pi * U1)
    Y = mirror_radius * np.sqrt(U2) * np.sin(2 * np.pi * U1)

    Z = radius_of_curvature+(X**2+Y**2)/radius_of_curvature

    U = (radius_of_curvature+offset_from_focus)*X/Z - X

    T = (np.abs(U/line_width)+0.5).astype(int)
    Tmask = T % 2 == 1
    XP = X0+X*C/mirror_diameter
    YP = Y0+Y*K*C/mirror_diameter

    xp_mask = np.ma.masked_where(Tmask, XP)
    yp_mask = np.ma.masked_where(Tmask, YP)

    plt.plot(xp_mask, yp_mask, 'o', markersize=0.1, color='blue')
    plt.xticks([])
    plt.yticks([])
    plt.title("offset = %.1fmm"%offset)
plt.show()

end = time.time()
print(end - start)
````
